```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

import xgboost as xgb
print("xgb.__version__:", xgb.__version__)

from sklearn.model_selection import GroupKFold

# local modules
import sys
sys.path.append("../src")
from preproc import preprocess_data
```

***
### load and preprocess data

```python
# define some paths
path_raw = Path("../data/raw")
path_processed = Path("../data/processed")

# load data
df_train = pd.read_csv(path_raw / "train.csv")
df_test = pd.read_csv(path_raw / "test.csv")

df_train
```

```python
# Call the function
df_train, df_test, numerical_cols, categorical_cols = preprocess_data(
    df_train,
    df_test,
    scale_utility=True
)

# Print the results
print("Numerical Columns:", len(numerical_cols))
print("Categorical Columns:", len(categorical_cols))
```

```python
total_games = df_train["num_wins_agent1"] + df_train["num_draws_agent1"] + df_train["num_losses_agent1"]
df_train["win_rate_agent1"] = df_train["num_wins_agent1"] / total_games
df_train["draw_rate_agent1"] = df_train["num_draws_agent1"] / total_games
df_train["loss_rate_agent1"] = df_train["num_losses_agent1"] / total_games


df_train["utility1"] = df_train.eval("win_rate_agent1 - loss_rate_agent1")
df_train["utility2"] = df_train.eval("win_rate_agent1 - draw_rate_agent1")
df_train["utility3"] = df_train.eval("loss_rate_agent1 - draw_rate_agent1")
```

```python
targets = [
    "win_rate_agent1", "draw_rate_agent1", "loss_rate_agent1",
    "utility1", "utility2", "utility3"
]

for target in targets:
    plt.hist(df_train[target], bins=100)
    plt.title(f'Distribution of {target}')
    plt.xlabel('Utility Agent 1')
    plt.ylabel('Frequency')
    plt.show()
```

***
### train model


```python
def custom_rmse0(predt: np.ndarray, dtrain: xgb.DMatrix):
    y_true = dtrain.get_label().reshape(predt.shape)
    y_true_utility = y_true[:,0]
    y_pred_utility = predt[:,0]
    rmse = np.sqrt(np.mean((y_pred_utility - y_true_utility) ** 2))
    return 'custom_rmse', rmse

def custom_rmse1(predt: np.ndarray, dtrain: xgb.DMatrix):
    y_true = dtrain.get_label().reshape(predt.shape)
    y_true_utility = y_true[:,1]
    y_pred_utility = predt[:,1]
    rmse = np.sqrt(np.mean((y_pred_utility - y_true_utility) ** 2))
    return 'custom_rmse', rmse

def custom_rmse2(predt: np.ndarray, dtrain: xgb.DMatrix):
    y_true = dtrain.get_label().reshape(predt.shape)
    y_true_utility = y_true[:,2]
    y_pred_utility = predt[:,2]
    rmse = np.sqrt(np.mean((y_pred_utility - y_true_utility) ** 2))
    return 'custom_rmse', rmse

# Define the number of folds for cross-validation
num_folds = 5

targets = ["utility1", "utility2", "utility3"]

# Define the column for stratified or group k-fold
groups_col = 'GameRulesetName'
gkf = GroupKFold(n_splits=num_folds)
split_list = gkf.split(df_train, groups=df_train[groups_col])

# Define the parameters
params = {
    'objective': "reg:squarederror",
    'eval_metric': 'rmse',
    'max_depth': 5,
    'learning_rate': 0.05,
    'multi_strategy': 'multi_output_tree',
    # 'disable_default_eval_metric': True,
}

trained_models = []
oof = pd.DataFrame(index=df_train.index, columns=['utility_agent1_true', 'utility_agent1_pred'])
oof_scores = []

# Perform cross-validation
for fold, (train_index, val_index) in enumerate(split_list, 1):
    print(f"Fold {fold}")
    
    # Split the data
    X_train, X_valid = df_train.iloc[train_index], df_train.iloc[val_index]
    y_train, y_valid = X_train[targets], X_valid[targets]

    feature_types = [
        "c" if col in categorical_cols else "q" 
        for col in numerical_cols + categorical_cols
    ]
    
    # Convert to DMatrix format
    train_data = xgb.DMatrix(
        data=X_train[numerical_cols + categorical_cols],
        label=y_train,
        feature_types=feature_types,
        enable_categorical=True,
    )
    val_data = xgb.DMatrix(
        data=X_valid[numerical_cols + categorical_cols],
        label=y_valid,
        feature_types=feature_types,
        enable_categorical=True,
    )

    # Train the model
    model = xgb.train(
        params,
        train_data,
        num_boost_round=2000,
        evals=[(val_data, 'val')],
        verbose_eval=25,
        custom_metric=custom_rmse2,
    )
    
    # Save the trained model to the list
    trained_models.append(model)

    # predict on validation set
    y_pred = model.predict(val_data)
    y_pred = np.clip(y_pred, -1, 1)
    
    # Compute RMSE on scaled values
    rmse = np.sqrt(np.mean((y_pred[:,0] - y_valid.iloc[:,0]) ** 2))
    print(f"Fold {fold} - RMSE: {rmse}")

    # Save out-of-fold predictions
    oof.loc[val_index, 'utility_agent1_true'] = y_valid.iloc[:,0]
    oof.loc[val_index, 'utility_agent1_pred'] = y_pred[:,0]

    # Save RMSE to the list
    oof_scores.append(rmse)

# Print the list of oof scores and average oof score
print("List of oof scores:", oof_scores)
print("Average oof score:", np.mean(oof_scores))
```

```python
plt.hist(oof['utility_agent1_true'], alpha=0.5, label='Target')
plt.hist(oof['utility_agent1_pred'], alpha=0.5, label='Prediction')
plt.legend()
plt.show()
```
